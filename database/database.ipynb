```python
import pandas as pd
```

```python
df = pd.read_csv('rximage_data.csv')
```

```python
df.head()
```

```python
import re

# Function to extract pill name
def extract_pill_name(name):
    if not isinstance(name, str):  # Ensure name is a string
        return name
    match = re.match(r"^(.*?)(?:\s\d+(\.\d+)?\s*(MG|MCG|G|Units|IU|ML|%|MEQ|MG\/ML))", name, re.IGNORECASE)
    return match.group(1).strip() if match else name  # Return extracted pill name

# Ensure 'name' column is treated as strings
df["name"] = df["name"].astype(str)

# Apply function to the 'name' column
df["pill_name"] = df["name"].apply(extract_pill_name)
```

```python
# Display updated DataFrame
print(df["pill_name"].sample(20))  # Shows the first few rows
```

```python

# Function to extract dosage
def extract_dosage(name):
    if not isinstance(name, str):  # Ensure name is a string
        return None
    match = re.search(r"(\d+(\.\d+)?)\s*(MG|MCG|G|Units|IU|ML|%|MEQ|MG\/ML)", name, re.IGNORECASE)
    return match.group(0) if match else None  # Return dosage if found

# Apply function to the 'name' column to extract dosage
df["dosage"] = df["name"].apply(extract_dosage)
```

```python
# Display updated DataFrame
print(df["dosage"].sample(20))  # Shows the first few rows
```

```python
df.columns
```

```python
# List of columns to drop
columns_to_drop = [
    'attribution',
    'ingredientsAvailable',
    'rxnavImageObjectId',
    'rxnavImageSize',
    'nlmImageObjectId',
    'nlmImageFileName',
    'nlmImageSize'
]

# Drop the columns
df = df.drop(columns=columns_to_drop, errors='ignore')  # 'ignore' prevents errors if a column is missing

# Display the updated DataFrame
print(df.columns)
```

```python
df.to_csv("pill_data.csv", index=False)
```
