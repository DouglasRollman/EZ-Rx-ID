```python
import pandas as pd
import os
import re

# Load the existing dataframe
df_rximages = pd.read_csv('rximage_data.csv')  # Update with actual path

# Set the directory containing the split images
image_folder = "D:/rximage/image/images/split"  # Update with the actual folder path

# Get all filenames in the directory
image_files = os.listdir(image_folder)

# Regular expression to match filenames ending with _top, _bottom, _left, _right
pattern = re.compile(r"^(.*)_(top|bottom|left|right)\.(jpg|png|jpeg)$")

# Create a list to store image data
image_data = []

# Scan images and classify them
for img in image_files:
    match = pattern.match(img)
    if match:
        base_name, split_side, ext = match.groups()
        split_type = "vertical" if split_side in ["top", "bottom"] else "horizontal"
        new_filename = f"{base_name}_{split_side}.jpg"
        image_data.append((base_name, new_filename, split_type, split_side))

# Convert the list to a DataFrame
df_split_info = pd.DataFrame(image_data, columns=["base_filename", "new_filename", "split_type", "split_side"])

# Add base filename column to the original dataframe for merging
df_rximages['base_filename'] = df_rximages['rxnavImageFileName'].apply(lambda x: os.path.splitext(x)[0])

# Merge this with the original dataframe
df_updated = df_rximages.merge(df_split_info, on="base_filename", how="left")

print("Dataframe updated with new filenames, split type, and side.")
```
